## Implementacja Stochastycznego Gradientu (Gradient Descent)

Użyjemy algorytmu stochastycznego spadku do znalezienia minimum funkcji straty określonej wzorem: $L(w) = w^{2}-4w $.

Pochodna tej funkcji to $\frac{dL}{dw}=2*w-4$

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

w = np.arange(-10, 14, 0.1)
Loss = w**2 - 4*w

point = -4
plt.figure(figsize=(8, 6))
plt.plot(point, point ** 2 -4 * point, 'ro')
plt.plot(w, Loss)
plt.title('Loss function')
```

```python
def gradient_descent(df=lambda w: 2* w - 4, learining_rate=0.01, w_0=-4, max_iters=10000, precision=0.000001):
    """
    parametry:
    ----------
    df: gradient funkcji do optymalizacji
    learning_rate: wskaźnik uczenia
    max_iters: maksymalna liczba iteracji
    precision: precyzja sprawdza kiedy zatrzymać działanie algorytmu
    w_0: inicjalizacja wagi początkowej
    """

    # licznik iteracji
    iters = 0

    # kontrola wartości kroku kolejnego spadku
    previous_step_size = 1

    weights = []

    while previous_step_size > precision and iters < max_iters:
        w_prev = w_0
        w_0 = w_0 - learining_rate * df(w_prev)
        previous_step_size = abs(w_0 - w_prev)
        iters += 1
        weights.append(w_0)
        print('Iter #{}: obecny punkt: {}'.format(iters, w_0))

    print('Minium globalne: {}'.format(w_0))
    return weights
```

```python
weights = gradient_descent(learining_rate=0.1)
```

```python
plt.figure(figsize=(8,6))
plt.plot(w, Loss)
for weight in weights:
    plt.plot(weight, weight ** 2 - 4 * weight, 'ro')
```
